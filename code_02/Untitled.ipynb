```python
import os
import numpy as np
from PIL import Image
```

```python
# Replace with your actual directory containing raw files
raw_dir = 'C:\\Users\\youga\\Desktop\\BTP\\copy_500'
raw_dir
```

```python
# Replace with the dimensions of your images
image_width = 1024
image_height = 384
```

```python
# Specify the output directory for PNG images
output_dir = 'C:\\Users\\youga\\Desktop\\BTP\\output'
# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python
# Loop through each raw file in the directory
for filename in os.listdir(raw_dir):
    if filename.endswith('.bin'):
        raw_filepath = os.path.join(raw_dir, filename)
        
        with open(raw_filepath, 'rb') as f:
            # Read raw data (assumes uint16 for 14-bit)
            raw_data = np.fromfile(f, dtype=np.uint16)
        
        # Reshape and normalize raw data
        image_data = raw_data.reshape(image_height, image_width)
        image_data = (image_data / 8191.0) * 255.0
        
        # Convert to uint8 and create PIL image
        image_data = image_data.astype(np.uint8)
        image = Image.fromarray(image_data, 'L')  # 'L' mode for grayscale
        
        # Save as PNG in the specified output directory
        png_filename = os.path.splitext(filename)[0] + '.png'
        png_filepath = os.path.join(output_dir, png_filename)
        image.save(png_filepath)
        
        print(f'Converted {filename} to {png_filename}')
```
