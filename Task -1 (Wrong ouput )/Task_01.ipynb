```python
import os
import numpy as np
from PIL import Image
```

```python
input_folder = "C:\\Users\\youga\\Desktop\\BTP\\copy_500"
output_folder = "C:\\Users\\youga\\Desktop\\BTP\\output"
```

```python
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
image_width = 512
image_height = 768
bits_per_pixel = 8
```

```python
for filename in os.listdir(input_folder):
    input_file_path = os.path.join(input_folder, filename)
    
    with open(input_file_path, "rb") as input_file:
        # Read the binary data
        binary_data = input_file.read()
        
        # Interpret binary data as 16-bit integers (assuming little-endian)
        pixel_values = np.frombuffer(binary_data, dtype=np.uint16)
        
        # Reshape the pixel values into a 2D array
        pixel_matrix = pixel_values.reshape(image_height, image_width)
        
        # Normalize pixel values to 8-bit grayscale range (0-255)
        normalized_pixel_matrix = (pixel_matrix * 255) // ((1 << bits_per_pixel) - 1)
        
        # Convert the numpy array to a PIL Image
        grayscale_image = Image.fromarray(normalized_pixel_matrix.astype(np.uint8), "L")
        
        # Save the image as PNG
        output_file_path = os.path.join(output_folder, filename + ".png")
        grayscale_image.save(output_file_path)
        
        print(f"Converted {filename} to PNG")
```

```python
print("Conversion completed.")
```
